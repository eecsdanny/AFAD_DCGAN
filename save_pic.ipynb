# save single picture for every model 




```
def plot_generated_images(generator,noise,path_save=None,titleadd=''):
    imgs = generator.predict(noise)
    fig = plt.figure(figsize=(40,10))
    for i, img in enumerate(imgs):
        ax = fig.add_subplot(1,5,i+1)
        ax.imshow(img)
    fig.suptitle("Generated images "+titleadd,fontsize=30)
    
    if path_save is not None:
        plt.savefig(path_save+titleadd+'.png',
                    bbox_inches='tight',
                    pad_inches=0)
        plt.show()

        plt.close()
    else:
        plt.show()
    del fig
    del imgs
    del ax
noise = np.random.normal(0,1,(5,100))
```

```
# /content/drive/My Drive/dcgan_models/
dir_base = '/content/drive/My Drive/DCGAN/model_history1205/'
for epoch in range(25):
    file_name = 'generator＿' + str(epoch) + '_'+'1000'
    g_path = dir_base + file_name +'.h5'
    generator = tf.keras.models.load_model(g_path)
    plot_generated_images(generator,noise,titleadd=file_name,path_save='/content/drive/My Drive/DCGAN/pictures_for_gif_1205/')
    del generator
```

# Save imgs for all models

```
def gen_pics(generator,num=1):
    noise = np.random.normal(0,1,(num,100))
    imgs = generator.predict(noise)
    return imgs
```

```
import os
import matplotlib

dir_base = '/content/drive/My Drive/DCGAN/model_history1205/'
save_dir_base ='/content/drive/My Drive/DCGAN/database2/'
for epoch in range(23):
  file_name = 'generator＿' + str(epoch+3) + '_'+'1000'
  g_path = dir_base + file_name +'.h5'
  generator = tf.keras.models.load_model(g_path)
  imgs = gen_pics(generator,1000)
  for i,img in enumerate(imgs):
    matplotlib.image.imsave(save_dir_base+str(epoch)+str(i)+'.png',img)
    print(epoch,i)
```
