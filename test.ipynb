# Generate and Plot 

```
import matplotlib.pyplot as plt
import numpy as np
# plot function
def plot_generated_images(generator, num_of_instance = 5):
  noise = np.random.normal(0,1,(num_of_instance,100))
  imgs = generator.predict(noise)
  row = (num_of_instance+4)/5
  plt.figure('Generated images',(10,10))
  for index, img in enumerate(imgs):
    plt.subplot(row,5,index+1) 
    plt.imshow(img)
  plt.show()
# modify second parameter to change number of instance generated
plot_generated_images(generator,8)
```
